```python
retail = retail.drop(['Postal Code'],axis = 1)
retail.head()
```

```python
import seaborn as sns
sns.pairplot(retail)
```

```python
retail.corr()
plt.figure(figsize=(10,10))
sns.heatmap(retail.corr(), annot=True)
```

```python
plt.figure(figsize = (6,6)).patch.set_facecolor('red')
plt.title('Category')
plt.pie(retail['Category'].value_counts(), labels=retail['Category'].value_counts().index,autopct='%1.1f%%')
plt.show()
```

```python
retail.groupby('Category')['Profit','Sales'].agg(['sum']).plot.bar()
plt.title('Sales and Profits')
plt.titlesize : 100
plt.ylabel('Profits')
plt.show()
```

```python
retail_sorted_desc= retail.sort_values('Profit',ascending=True)

retail_sorted_desc.groupby('Sub-Category')['Profit','Sales'].agg(['sum']).plot.barh(figsize = (10,10))
plt.title('Sales and Profits')
plt.titlesize = 100
plt.ylabel('Profits and sales')
plt.show()
```

```python

retail.groupby(['Sub-Category'])['Profit'].sum().nlargest(100).plot.barh(figsize=(5,5)) 
```

```python
sns.set_theme(style="whitegrid")
plt.figure(figsize=(6,6))
sns.lineplot(retail['Discount'], retail['Profit'], data = retail)
```

```python
sns.set_theme(style="whitegrid")
top_states_s = retail.groupby("State").Sales.sum().nlargest(n=10)
top_states_p = retail.groupby("State").Profit.sum().nlargest(n = 10)
top_states_s.plot(kind = 'bar',figsize = (10,5),fontsize = 14)
top_states_p.plot(kind = 'bar',figsize = (10,5),fontsize = 14, color = 'red')
plt.xlabel('States')
plt.ylabel('Total sales')
#sns.barplot(x="Profit", y="State", data = top_states_p, label= "Profit" , color="b")
plt.title("Top 10 states Sales vs Profit")
```

```python
sns.set_theme(style="whitegrid")
top_states_s = retail.groupby("State").Sales.sum().nsmallest(n=10)
top_states_p = retail.groupby("State").Profit.sum().nsmallest(n = 10)
top_states_s.plot(kind = 'bar',figsize = (10,5),fontsize = 14)
top_states_p.plot(kind = 'bar',figsize = (10,5),fontsize = 14, color = 'red')
plt.xlabel('States')
plt.ylabel('Total sales')
#sns.barplot(x="Profit", y="State", data = top_states_p, label= "Profit" , color="b")
plt.title("least 10 states Sales vs Profit")
```

```python
plt.style.use('seaborn')
retail.plot(kind = "scatter",figsize = (10,5), x = "Sales", y= "Profit", c = "Discount", s = 20, colormap = "viridis")
plt.ylabel('Total Profits')
plt.xlabel('Sales')
plt.title("Interdependency of Sales,Profits and Discounts")
```
